```python
survived = 'survived'
not_survived = 'not_survived'
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(32,16))
women = titanic[titanic['Sex']=='female']
men = titanic[titanic['Sex']=='male']
plt.grid()
ax = sns.distplot(women[women['Survived']==1].Age.dropna(),bins=18,label=survived,ax=axes[0],kde=False)
ax = sns.distplot(women[women['Survived']==0].Age.dropna(),bins=40,label=not_survived,ax=axes[0],kde=False)
ax.legend()

ax.set_title('Female')
ax = sns.distplot(men[men['Survived']==1].Age.dropna(),bins=36,label=survived,ax=axes[1],kde=False)
ax = sns.distplot(men[men['Survived']==0].Age.dropna(),bins=40,label=not_survived,ax=axes[1],kde=False)
ax.legend()
_ = ax.set_title('Male')
```

# influnce of passenger class on survial

```python
sns.barplot(x='Pclass',y='Survived',data=titanic,palette=['green','yellow','red'])
plt.rc('xtick',labelsize=14)
plt.rc('ytick',labelsize=14)
plt.figure()
fig = titanic.groupby('Survived')['Pclass'].plot.hist(histtype='bar',alpha=0.8)
plt.legend(('Died','Survived'),fontsize=12)
plt.xlabel('Pclass',fontsize=20)
plt.show()
```
